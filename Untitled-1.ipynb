```python
def remove_sql_comments(sql_code):
    # Loại bỏ tất cả các comment dạng /* */ kể cả trên nhiều dòng
    sql_code_no_block_comments = re.sub(r'/\*.*?\*/', '', sql_code, flags=re.DOTALL)
    
    # Tách đoạn mã SQL thành từng dòng
    lines = sql_code_no_block_comments.splitlines()
    
    # Xử lý từng dòng để loại bỏ comment dạng --
    cleaned_lines = []
    for line in lines:
        # Tìm vị trí ký tự comment --
        comment_index = line.find('--')
        if comment_index != -1:
            # Nếu có comment, chỉ lấy phần trước comment
            line = line[:comment_index]
        
        # Thêm dòng đã được làm sạch vào danh sách, chỉ thêm nếu dòng không trống
        if line.strip():
            cleaned_lines.append(line.strip())
    
    # Nối các dòng lại với nhau, mỗi dòng cách nhau bởi dấu xuống dòng
    return '\n'.join(cleaned_lines)
```
